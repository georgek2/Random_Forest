```python
import pandas as pd

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import cross_val_score
```

```python
melb = pd.read_csv('melb_data.csv')

melb.head()
```

```python
# Separating the target from the features

y = melb.Price

X = melb.drop(['Price'], axis=1)
```

```python
# Selecting the numerical and categorical features to use

numerical_cols = [col for col in X.columns if X[col].dtype in ['int64', 'float64'] and X[col].isnull().sum() < 100]

categorical_cols = [col for col in X.columns if X[col].dtype == 'object' and X[col].nunique() < 10]
```

```python
# Setting the transformers to impute and encode the feature columns

numerical_transformer = SimpleImputer()

categorical_transformer = OneHotEncoder(handle_unknown='ignore', sparse=False)
```

```python
# The preprocessor to transform feature columns

preprocessor = ColumnTransformer(transformers=[('num', numerical_transformer, numerical_cols), 
                                               ('cat', categorical_transformer, categorical_cols)])
```

```python
# Splitting the training and the validation sets

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# The model

model = RandomForestRegressor(random_state=0)

# The pipeline that bundles the preprocessing and modelling

my_pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('model', model)])
```

```python
# The scores from Cross_Validation
# Five Folds

scores = -1 * cross_val_score(my_pipeline, X, y, cv=5, scoring='neg_mean_absolute_error')

scores
```

```python
# The mean score from all the folds

scores.mean()
```
